```python
import tensorflow as tf
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
temperature =[0,1,2,3,4,5]
#normal softmax function
def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a/sum_exp_a
    return y
q = softmax(temperature)
# softmax with temperature paramater

def softmax_with_temperature(temperature, T):
    temperature = np.array(temperature)
    temperature = temperature/T
    max_temperature = np.max(temperature)
    exp_temperature = np.exp(temperature -max_temperature)
    sum_exp_temperature = np.sum(exp_temperature)
    y = exp_temperature/sum_exp_temperature
    return y
calibrated_q1 = softmax_with_temperature(temperature, 0.01)
calibrated_q2 = softmax_with_temperature(temperature, 0.2)
calibrated_q3 = softmax_with_temperature(temperature, 0.5)
calibrated_q4 = softmax_with_temperature(temperature, 1.2)

fig = sns.barplot(x=np.array(temperature), y=q)
fig.set(ylim=(0,1))
plt.show()

fig2 = sns.barplot(x=np.array(temperature), y=calibrated_q1)
fig2.set(ylim=(0,1))
plt.show()

fig3 = sns.barplot(x=np.array(temperature), y=calibrated_q2)
fig3.set(ylim=(0,1))
plt.show()

fig4 = sns.barplot(x=np.array(temperature), y=calibrated_q3)
fig4.set(ylim=(0,1))
plt.show()

fig5 = sns.barplot(x=np.array(temperature), y=calibrated_q4)
fig5.set(ylim=(0,1))
plt.show()
```

### We can see thet as the size of temperature decreases, the difference between probability  decreases. But, there is no change in the order so it will not give any affects on the accuracy
